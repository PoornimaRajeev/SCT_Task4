```python
import pandas as pd

# Load the Excel file
df = pd.read_csv("road_accident_data.csv")

```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract day of week
df['Day'] = df['Date'].dt.day_name()

# Remove leading/trailing spaces from column values
for col in ['Time_of_Day', 'Weather_Conditions', 'Road_Conditions', 'Light_Conditions']:
    df[col] = df[col].astype(str).str.strip()

# Drop rows with missing key values
df.dropna(subset=['Time_of_Day', 'Weather_Conditions', 'Road_Conditions', 'Light_Conditions'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.countplot(
    y='Road_Conditions',
    data=df,
    order=df['Road_Conditions'].value_counts().index,
    palette="coolwarm"
)
plt.title("Accidents by Road Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Condition")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='Weather_Conditions', data=df, order=df['Weather_Conditions'].value_counts().index, palette="pastel")
plt.title("Accidents by Weather Condition")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.countplot(x='Light_Conditions', data=df, order=df['Light_Conditions'].value_counts().index, palette="muted")
plt.title("Accidents by Light Conditions")
plt.grid(True)
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(y='Road_Conditions', data=df, order=df['Road_Conditions'].value_counts().index, palette="coolwarm")
plt.title("Accidents by Road Conditions")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='Day', data=df, order=order, palette="Set2")
plt.title("Accidents by Day of the Week")
plt.grid(True)
plt.show()
```
